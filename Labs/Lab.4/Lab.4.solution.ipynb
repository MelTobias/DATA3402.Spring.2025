1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value):
        self.max_value=max_value
        self.count=0
        
    def increment(self):
        if self.count<self.max_value:
            self.count+=1
        else:
            print(f"Error: cannot increment past {self.max_value}")
            
    def reset(self):
        self.count=0
        
    def get_count(self):
        return self.count
```

```python
#test

#makes a Counter with the max_value of 3
counter=Counter(3)

counter.increment()
print(counter.get_count()) #output should be 1

counter.increment()
print(counter.get_count()) #output should be 2

counter.increment()
print(counter.get_count()) #output should be 3

counter.increment() #should be error output

#resets the counter
counter.reset()
print(counter.get_count()) #output should be 0
```

```python
class Shape:
    def area(self):
        raise NotImplementedError("Subclass must implement this method")

    def perimeter(self):
        raise NotImplementedError("Subclass must implement this method")

    def get_coordinates(self):
        raise NotImplementedError("Subclass must implement this method")

    def perimeter_points(self, num_points=16):
        raise NotImplementedError("Subclass must implement this method") 

    def is_inside(self, x, y):
        raise NotImplementedError("Subclass must implement this method")

    def overlaps_with(self, other):
        raise NotImplementedError("Subclass must implement this method")

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length=length
        self.__width=width
        self.__x=x
        self.__y=y

    def area(self):
        return self.__length*self.__width

    def perimeter(self):
        return 2*(self.__length+self.__width)

    def get_coordinates(self):
        return (self.__x, self.__y)

    def perimeter_points(self, num_points=16):
        points=[]
        for i in range(num_points):
            if i<num_points//4:
                x=self.__x+(self.__length*i/(num_points//4))
                y=self.__y
            elif i<num_points//2:
                x=self.__x+self.__length
                y=self.__y+(self.__width*(i-num_points//4)/(num_points//4))
            elif i<3*num_points//4:
                x=self.__x+self.__length-(self.__length*(i-num_points//2)/(num_points//4))
                y=self.__y+self.__width
            else:
                x=self.__x
                y=self.__y+self.__width-(self.__width*(i-3*num_points//4)/(num_points//4))
            points.append((x, y))
        return points

    def is_inside(self, x, y):
        return self.__x<=x<=self.__x+self.__length and self.__y<=y<=self.__y+self.__width

    #checks if projections overlap
    def overlaps_with(self, other):
        if isinstance(other, Rectangle):
            return not (self.__x+self.__length<other.__x or self.__x>other.__x+other.__length or
                        self.__y+self.__width<other.__y or self.__y>other.__y+other.__width)
        return False

class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius=radius
        self.__x=x
        self.__y=y

    def area(self):
        return math.pi*(self.__radius**2)

    def perimeter(self):
        return 2*math.pi*self.__radius
        
    def get_coordinates(self):
        return (self.__x, self.__y)

    def perimeter_points(self, num_points=16):
        points=[]
        for i in range(num_points):
            angle=2*math.pi*i/num_points
            x=self.__x+self.__radius*math.cos(angle)
            y=self.__y+self.__radius*math.sin(angle)
            points.append((x, y))
        return points

    def is_inside(self, x, y):
        distance=math.sqrt((x-self.__x)**2+(y-self.__y)**2)
        return distance<=self.__radius

    #checks if the distance between the centers is less than the sum of the radius
    def overlaps_with(self, other):
        if isinstance(other, Circle):
            distance=math.sqrt((self.__x-other.__x)**2+(self.__y-other.__y)**2)
            return distance<(self.__radius+other.__radius)
        return False

class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        self.__side1=side1
        self.__side2=side2
        self.__side3=side3
        self.__x=x
        self.__y=y

    def area(self):
        s=(self.__side1+self.__side2+self.__side3)/2
        return math.sqrt(s*(s-self.__side1)*(s-self.__side2)*(s-self.__side3))

    def perimeter(self):
        return self.__side1+self.__side2+self.__side3

    def get_coordinates(self):
        return (self.__x, self.__y)

    def get_sides(self):
        return (self.__side1, self.__side2, self.__side3)

    def perimeter_points(self, num_points=16):
        points=[]
        points_per_side=num_points//3
        for i in range(points_per_side):
            #side 1
            x=self.__x+(self.__side1*i/points_per_side)
            y=self.__y
            points.append((x, y))
    
            #side 2
            x=self.__x+self.__side1
            y=self.__y+(self.__side2*i/points_per_side)
            points.append((x, y))
    
            #side 3
            x=self.__x+self.__side1-(self.__side1*i/points_per_side)
            y=self.__y+self.__side2
            points.append((x, y))
        return points[:num_points]
        
    def is_inside(self, x, y):
        def sign(p1, p2, p3):
            return (p1[0]-p3[0])*(p2[1]-p3[1])-(p2[0]-p3[0])*(p1[1]-p3[1])
        p1=(self.__x, self.__y)
        p2=(self.__x+self.__side1, self.__y)
        p3=(self.__x+self.__side1, self.__y+self.__side2)

        d1=sign((x, y), p1, p2)
        d2=sign((x, y), p2, p3)
        d3=sign((x, y), p3, p1)

        has_neg=(d1<0) or (d2<0) or (d2<0)
        has_pos=(d1>0) or (d2>0) or (d3>0)
        return not (has_neg and has_pos)

    #checks if any vertex is inside another
    def overlaps_with(self, other):
        if isinstance(other, Triangle):
            return (self.is_inside(other.__x, other.__y) or
                    other.is_inside(self.__x, self.__y))
        return False
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(y, y + w):
            self.set_pixel(i, x, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(x, x + h):
            self.set_pixel(y, i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        # Using Bresenham's line algorithm for better accuracy
        dx = abs(x2 - x1)
        dy = abs(y2 - y1)
        sx = 1 if x1 < x2 else -1
        sy = 1 if y1 < y2 else -1
        err = dx - dy
        
        while True:
            self.set_pixel(y1, x1, **kargs)
            if x1 == x2 and y1 == y2:
                break
            e2 = err * 2
            if e2 > -dy:
                err -= dy
                x1 += sx
            if e2 < dx:
                err += dx
                y1 += sy

    def display(self):
        print("\n".join(["".join(row) for row in self.data]))

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length=length
        self.__width=width
        self.__x=x
        self.__y=y

    def area(self):
        return self.__length*self.__width

    def perimeter(self):
        return 2*(self.__length+self.__width)

    def get_coordinates(self):
        return (self.__x, self.__y)

    def perimeter_points(self, num_points=16):
        points=[]
        for i in range(num_points):
            if i<num_points//4:
                x=self.__x+(self.__length*i/(num_points//4))
                y=self.__y
            elif i<num_points//2:
                x=self.__x+self.__length
                y=self.__y+(self.__width*(i-num_points//4)/(num_points//4))
            elif i<3*num_points//4:
                x=self.__x+self.__length-(self.__length*(i-num_points//2)/(num_points//4))
                y=self.__y+self.__width
            else:
                x=self.__x
                y=self.__y+self.__width-(self.__width*(i-3*num_points//4)/(num_points//4))
            points.append((x, y))
        return points

    def is_inside(self, x, y):
        return self.__x<=x<=self.__x+self.__length and self.__y<=y<=self.__y+self.__width

    def overlaps_with(self, other):
        if isinstance(other, Rectangle):
            return not (self.__x+self.__length<other.__x or self.__x>other.__x+other.__length or
                        self.__y+self.__width<other.__y or self.__y>other.__y+other.__width)
        return False

    def paint(self, canvas, char='*'):
        # horizontal lines
        canvas.h_line(self.__x, self.__y, self.__length, char=char)
        canvas.h_line(self.__x, self.__y+self.__width, self.__length, char=char)
        # vertical lines
        canvas.v_line(self.__x, self.__y, self.__width, char=char) 
        canvas.v_line(self.__x+self.__length, self.__y, self.__width, char=char) 

class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius=radius
        self.__x=x
        self.__y=y

    def area(self):
        return math.pi*(self.__radius**2)

    def perimeter(self):
        return 2*math.pi*self.__radius
        
    def get_coordinates(self):
        return (self.__x, self.__y)

    def perimeter_points(self, num_points=16):
        points=[]
        for i in range(num_points):
            angle=2*math.pi*i/num_points
            x=self.__x+self.__radius*math.cos(angle)
            y=self.__y+self.__radius*math.sin(angle)
            points.append((x, y))
        return points

    def is_inside(self, x, y):
        distance=math.sqrt((x-self.__x)**2+(y-self.__y)**2)
        return distance<=self.__radius

    def overlaps_with(self, other):
        if isinstance(other, Circle):
            distance=math.sqrt((self.__x-other.__x)**2+(self.__y-other.__y)**2)
            return distance<(self.__radius+other.__radius)
        return False

    def paint(self, canvas, char='*'):
        for angle in range(0, 360, 15):
            rad=math.radians(angle)
            x=int(self.__x+self.__radius*math.cos(rad))
            y=int(self.__y+self.__radius*math.sin(rad))

            if 0<=x<canvas.width and 0<=y<canvas.height:
                canvas.set_pixel(y, x, char)

class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        self.__side1=side1
        self.__side2=side2
        self.__side3=side3
        self.__x=x
        self.__y=y

    def area(self):
        s=(self.__side1+self.__side2+self.__side3)/2
        return math.sqrt(s*(s-self.__side1)*(s-self.__side2)*(s-self.__side3))

    def perimeter(self):
        return self.__side1+self.__side2+self.__side3

    def get_coordinates(self):
        return (self.__x, self.__y)

    def perimeter_points(self, num_points=16):
        points=[]
        points_per_side=num_points//3
        for i in range(points_per_side):
            x=self.__x+(self.__side1*i/points_per_side)
            y=self.__y
            points.append((x, y))

            x=self.__x+self.__side1
            y=self.__y+(self.__side2*i/points_per_side)
            points.append((x, y))

            x=self.__x+self.__side1-(self.__side1*i/points_per_side)
            y=self.__y+self.__side2
            points.append((x, y))
        return points[:num_points]
        
    def is_inside(self, x, y):
        def sign(p1, p2, p3):
            return (p1[0]-p3[0])*(p2[1]-p3[1])-(p2[0]-p3[0])*(p1[1]-p3[1])
        p1=(self.__x, self.__y)
        p2=(self.__x+self.__side1, self.__y)
        p3=(self.__x+self.__side1, self.__y+self.__side2)

        d1=sign((x, y), p1, p2)
        d2=sign((x, y), p2, p3)
        d3=sign((x, y), p3, p1)

        has_neg=(d1<0) or (d2<0) or (d3<0)
        has_pos=(d1>0) or (d2>0) or (d3>0)
        return not (has_neg and has_pos)

    def overlaps_with(self, other):
        if isinstance(other, Triangle):
            return (self.is_inside(other.__x, other.__y) or
                    other.is_inside(self.__x, self.__y))
        return False

    def paint(self, canvas, char='*'):
        x2=self.__x+self.__side1
        y2=self.__y
        x3=self.__x+self.__side1//2
        y3=self.__y-self.__side2

        if 0<=x2<canvas.width and 0<=y2<canvas.height:
            canvas.line(self.__x, self.__y, x2, y2, char=char)
        if 0<=x3<canvas.width and 0<=y3<canvas.height:
            canvas.line(x2, y2, x3, y3, char=char)
        if 0<=self.__x< canvas.width and 0<=self.__y<canvas.height:
            canvas.line(x3, y3, self.__x, self.__y, char=char)

class CompoundShape(Shape):
    def __init__(self):
        self.shapes=[]

    def add_shape(self, shape):
        self.shapes.append(shape)

    def area(self):
        return sum(shape.area() for shape in self.shapes)

    def perimeter(self):
        return sum(shape.perimeter() for shape in self.shapes)

    def get_coordinates(self):
        return [shape.get_coordinates() for shape in self.shapes]

    def perimeter_points(self, num_points=16):
        points=[]
        for shape in self.shapes:
            points.extend(shape.perimeter_points(num_points))
        return points

    def is_inside(self, x, y):
        return any(shape.is_inside(x, y) for shape in self.shapes)

    def overlaps_with(self, other):
        if isinstance(other, CompoundShape):
            return any(shape1.overlaps_with(shape2) for shape1 in self.shapes for shape2 in other.shapes)
        else:
            return any(shape.overlaps_with(other) for shape in self.shapes)

    def paint(self, canvas, char='*'):
        for shape in self.shapes:
            shape.paint(canvas, char)
```
